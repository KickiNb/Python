### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

### The explicit and the implicit interfaces
**The Object-Oriented approach:** Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style"). Useful for more complex visualizations or when you need fine-grained control over the plot elements. Recommended as it gives more control, detalied and better to customize your plots.

**The Pyplot-approach / MATLAB:** Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting. It's a simple and convenient way to quickly create plots, especially for interactive use and scripts

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

#### Figure
Figure is seen as an empty canvas before your start drawing. It keeps track of all the child **Axes**, a group of 'special **Artists** (like titles, figure legends, colorbars, etc) 
#### Axes
The part of the canvas that the plot or a graph is attached on. You call methods that do the plotting directly from the Axes which gives high flexibility in costumizing your plots. 

#### Axis
The marks on the Axis, (x- and y-axis) labeld as ticks. The location of the ticks is determined by a Locator object, and the tickabel stings formatted by a Formatter.

#### Artist
Everything that we see on the figure, the visible on the Figure is an Artist, including, figure, axes, and axis objects. 


### When plotting in Matplotlib, what is the expected input data type?

Plotting functions expect numpy.array or numpy.ma.masked_array as input, or objects that can be passed to numpy.asarray.


### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# The Object-Oriented-approach
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()

ax.plot(x, y, 'b:', label = 'quadratic')
ax.plot(x, x, 'g--', label = 'linear')


ax.set_xlabel('x')  
ax.set_ylabel('y')  
ax.set_title("Our Object-oriented approach x^2")  

ax.legend();
```

```python
# The Pyplot-approach:

x = np.linspace(-4, 4, 100)  
y = x**2

plt.plot(x, y, 'b:', label = 'quadratic')
plt.plot(x, x, 'g--', label = 'linear')

plt.title('Pyplot x^2')
plt.xlabel('x')
plt.ylabel('y')

plt.legend();
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

```

```python
# Scatter plot:

fig, axs = plt.subplots(2, 1, figsize =(10, 10))

axs[0].scatter(random_data_x, random_data_y, alpha=0.5)

axs[0].plot(x, y, color='red')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Scatter Plot with x^2')


fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

axs[1].bar(names, values, color=['purple', 'green', 'orange', 'yellow', 'blue'])
axs[1].set_title('Bar Plot of Fruit Data')
axs[1].set_xlabel('Fruit')
axs[1].set_ylabel('Quantity')
```
